# 3

```python
def find_inv(z):
    res = []
    for x in range(-z, z + 1):
        for y in range(-z, z + 1):
            if (x * y) % z == 1:
                res.append((x, y, x*y%z))
    return res
```

```python
find_inv(36)
```

```python
from sympy import mod_inverse
for x in range(1, 36):
    try:
        print(mod_inverse(x, 36))
    except:
        pass
```
