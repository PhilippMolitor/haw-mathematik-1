# 6. Mathematisches summieren mit Sigma ($\Sigma$)

```python
# s muss eine funktion sein mit genau einem parameter
def newsum(s, a, b):
    result = 0
    counter = a
    
    while counter <= b:
        result += s(counter) # funktion s mit einem parameter
        counter += 1

    return result

# beispiel f√ºr s:
# identische abbildung (keine ver√§nderung)
def id(x):
    return x

# beispiel f√ºr s:
# quadrieren (y*y)
def square(y):
    return y**2

newsum(id, 1, 100), newsum(square, 1, 100)
```

# 7. Gau√üsche Summenformel

$\sum_{i=1}^{n} i = \frac{n(n+1)}{2}$

f√ºr gerade $n$: $\frac{n}{2}(n+1)$

f√ºr ungerade $n$: $\frac{n-1}{2}(n+1)+(\frac{n+1}{2}) = \frac{n(n+1)}{2}$

```python
# Iterativ, rechenaufw√§ndig
def gaussian_sum_expensive(n):
    result = 0
    for i in range(n + 1):
        result += i
    return result

# Gek√ºrzt, weniger rechenaufw√§ndig
def gaussian_sum(n):
    if n % 2 == 0:
        return (n/2)*(n+1)
    else:
        return n*(n+1)/2

(
    gaussian_sum_expensive(10), gaussian_sum(10), # 10
    gaussian_sum_expensive(11), gaussian_sum(11), # 11
    gaussian_sum_expensive(100), gaussian_sum(100) # 100
)
```

# 9. Summieren von $a$ bis $b$ (statt $1$ bis $b$)

$\sum_{k=10}^{100} k = \sum_{k=1}^{100} k - \sum_{k=1}^{9} k$

$\sum_{k=a}^{b} S_{k} = \sum_{k=a}^{b} S_{k} - \sum_{k=a}^{a-1} S_{k}$

```python
def range_sum(a, b):
    sum_big = gaussian_sum(b)
    sum_small = gaussian_sum(a - 1)
    
    return sum_big - sum_small

range_sum(5, 10), range_sum(50, 100), range_sum(3, 9)
```

# 11. Geometrische Summenformel

$\sum_{i=0}^{5} 3^{i} = 3^{0} + 3^{1} + 3^{2} + 3^{3} + 3^{4} + 3^{5} = S$

$3 * S = 3^{1} + 3^{2} + 3^{3} + 3^{4} + 3^{5}+ 3^{6}$

$S = \frac{1-3^{6}}{1-3}$

$\sum_{i=0}^{n} q^{i} = \frac{1-q^{n+1}}{1-q}$

```python
# Rechenaufw√§ndig, manuell
def geometric_sum_expensive(n, q, i):
    result = 0
    for i in range(n + 1):
        result += q ** i
    return result

# Gek√ºrzte variante, weniger rechenaufw√§ndig
def geometric_sum(n, q, i):
    return (1 - (q ** (n + 1))) / (1 - q)

geometric_sum_expensive(4, 3, 2), geometric_sum(4, 3, 2)
```

# 12. Das Produktzeichen

$U_{1} * U_{2} = \prod_{i=1}^{2} U_{i}$

$\prod_{i=1}^{n} i = 1*2*3*...*n$ (Fakult√§t)

```python
def facul(n):
    return n * facul(n - 1) if n > 1 else 1

facul(1), facul(5), facul(10)
```
